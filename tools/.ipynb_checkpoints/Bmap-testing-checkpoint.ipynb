### Testing B maps

Import packages that we'll use.

```python
import pandas
from scipy import interpolate
import numpy as np
import sys
import matplotlib.pylab as plt

from bmap_prediction import *
```

Import the data

```python
### You will need to change the relative paths
data = np.loadtxt("../steady_state/Bvalues.steady_state.N_1000.s_-0.01.u_1e-07.r_1e-07.txt")
data /= (4 * 1000 * 1e-7)
plt.plot(data)
```

```python
s = -0.001
u = 1e-8
```

Load the data table and set up the grid of selected elements and positions to compute B values for.

```python
### You will need to change the relative paths
df = pandas.read_csv("mpp_stats_demo.csv.gz")
df_sub = df[(df["Na"] == 10000) & (df["Generation"] == 0) & (df["uL"] == 1e-8) & (df["N1"]==1000)]
u_vals, s_vals, splines = generate_cubic_splines(df_sub)

# Set up chromosome
L = 1e5
L_elem = 500
x_elem = np.arange(L_elem / 2, L, L_elem)

# The grid of positions that we compute B values at
xx = np.linspace(0, L, len(data))
```

```python
n_iter = 3
Bmap = get_Bmap_uniform(xx, x_elem, L_elem, u_vals, s_vals, splines, s=s, u=u)

plt.plot(xx, data, "k--", label="data")
plt.plot(xx, Bmap(xx), label=0)
for i in range(n_iter):
    Bmap = get_Bmap_adjusted(xx, x_elem, L_elem, u_vals, s_vals, splines, Bmap, s=s, u=u)
    plt.plot(xx, Bmap(xx), label=i+1)

plt.legend(fontsize=6, ncols=2)
plt.xlabel("Position")
plt.ylabel("B value");
```
